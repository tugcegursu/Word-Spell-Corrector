```python
#Importing necessary libraries
import nltk
import collections
#Creating functionc to calculate bigram and trigrams
def bigram_counts(word_list):
    bgs = nltk.bigrams(word_list)
    fdist = nltk.FreqDist(bgs)
    d = collections.Counter()
    for k, v in fdist.items():
        d[k] = v
    return d
def trigram_counts(word_list):
    tgs = nltk.trigrams(word_list)
    fdist = nltk.FreqDist(tgs)
    d = collections.Counter()
    for k, v in fdist.items():
        d[k] = v
    return d 
```

```python
import re
from collections import Counter

def words(text): return re.findall(r'\w+', text.lower())

WORDS = Counter(words(open('corpus_finalproject.txt').read()))

#create a function to calculate the probability of words
def prob(word, N=sum(WORDS.values())): 
    return WORDS[word] / N

#create a function to correct the misspelled words
def correct(words): 
    word_list = []
    for word in words:
        spelled = max(candidates(word), key=prob)
        word_list.append(spelled)
    return word_list

#create a function to find possible correction for misspelled words
def candidates(word): 
    return (known_words([word]) or known_words(edits1(word)) or known_words(edits2(word)) or [word])

def known_words(words): 
    return set(w for w in words if w in WORDS)

def edits1(word):
    letters    = 'abcçdefgğhıijklmnoöprsştuüvyz'
    splits     = [(word[:i], word[i:])    for i in range(len(word) + 1)]
    deletes    = [L + R[1:]               for L, R in splits if R]
    transposes = [L + R[1] + R[0] + R[2:] for L, R in splits if len(R)>1]
    replaces   = [L + c + R[1:]           for L, R in splits if R for c in letters]
    inserts    = [L + c + R               for L, R in splits for c in letters]
    return set(deletes + transposes + replaces + inserts)

def edits2(word): 
    return (e2 for e1 in edits1(word) for e2 in edits1(e1))
```
